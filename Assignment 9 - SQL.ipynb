```python
# imports needed packages.
from bs4 import BeautifulSoup as bs
import requests
import sqlite3
import pandas as pd
import csv
```

```python
url="https://books.toscrape.com/"
response= requests.get(url)
soup=bs(response.text,'html.parser')
```

```python
# links.remove(links[0])
links= soup.find('ul', class_="nav nav-list").find_all('a')[1:]
links
```

```python
# create Books DB and  Categories table inside it.
conn=sqlite3.connect('Books.db')
conn.execute("create table categories (CategoryId INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT, Category text)")
conn.close()
```

```python
# open connection to DB to insert data to categories table.
conn=sqlite3.connect('Books.db')
categoryID =0
for link in links:
    url2= url+link.get('href')
    category= link.get_text().strip()
    categoryID +=1
    
    conn.execute("INSERT INTO categories (Category) VALUES (?)",[category])
    conn.commit()
    
conn.close()
    
```

```python
# create a books table inside DB.
# define the CategoryId as Foreign Key.
conn=sqlite3.connect('Books.db')
conn.execute("create table books (BookId INTEGER PRIMARY KEY AUTOINCREMENT, Title text, Price real, Rate int,Category_Id int)")
conn.close()
```

```python
# open connection to DB to insert data to books table.
conn=sqlite3.connect('Books.db')
categoryid =0
for link in links:
    url2= url+link.get('href')
    category= link.get_text().strip()
    categoryid +=1
    response2= requests.get(url2)
    soup2= bs(response2.text,"html.parser")
    books= soup2.find_all('article', attrs={'class':"product_pod"})
    print(category)

    
    nextt= soup2.find('li', attrs={'class':"next"})
    while nextt !=None:


        url3= url2.replace('index.html', nextt.find('a').get('href'))

        response3= requests.get(url3)
        soup3= bs(response3.text,"html.parser")
        books2= soup3.find_all('article', attrs={'class':"product_pod"})
        for b in books2:
            books.append(b)
    
        nextt= soup3.find('li', attrs={'class':"next"})


   

    for book in books:
            price= float((book.find('p', attrs={'class':'price_color'}).get_text()).replace('Â£',''))
            title= book.find('h3').find('a').get('title')
            stars = book.find('p').get('class')[1]
            rates= {'One':1,'Two':2,'Three':3,"Four":4,'Five':5}
            rating= int(rates[stars])
            print(title,price,rating, categoryid )
            
            query = "INSERT INTO books (Title, Price, Rate,Category_Id) VALUES (?,?,?,?)"
            values = [title,price,rating,categoryid]
            conn.execute(query, values)
            conn.commit()
    
conn.close()

```
